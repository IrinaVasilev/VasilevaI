```python
#Импортируем библиотеку и создаем пустые графы
import networkx as nx
G = nx.DiGraph()
G1 = nx.DiGraph()
```

```python
# Добавляем узлы с весами
G.add_weighted_edges_from([("1","2", 0.2), ("2","3",0.6) , ("1","3",0.9), 
                           ("2","4",0.8), ("3","4",0.1), ("3","5",0.3), ("4","5",0.4), 
                           ("4","6",0.35), ("6","7",0.5), ("5","7",0.25)])
```

```python
# Добавляем узлы без весов 
G1.add_edges_from([("A","B"), ("B","C") , ("A","C"), 
                   ("B","4"), ("C","4"), ("C","5"), 
                   ("4","5"), ("4","6"), ("6","7")])
```

```python
# Обозначаем позицию, ее можно потом изменить вручную
pos = nx.spring_layout(G)
```

```python
# Рисуем граф для G

# Добавляем обозначения в виде весов
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels)
# Рисуем сам граф
nx.draw(G, pos = pos, with_labels = True, font_weight = 'bold')
```

```python
# Обозначаем позицию круговую, ее можно потом изменить вручную
pos = nx.circular_layout(G1)
```

```python
nx.draw(G1, pos = pos, with_labels = True, font_weight = 'bold')
```
